```python
#What countries in Africa is causing the spike in TB deaths and cases? (limit 10)
df.query('Region == "AFR"').groupby('Country or territory name')\
         ['Estimated prevalence of TB (all forms) per 100 000 population'].agg(['mean','count']) \
    .sort_values(by='mean', ascending=False).iloc[0:10]
```

```python
df.query('Region == "AFR"').groupby('Country or territory name')\
         ['Estimated prevalence of TB (all forms) per 100 000 population'].agg(['mean']) \
     .sort_values('mean')['mean'] \
    .iloc[0:10] \
    .plot(kind='barh', figsize=(12, 5), title='Top 10 African Countries for Estimated prevalence of TB per 100k')
ax.set_xlabel('#cases', 
               fontweight ='bold')
plt.show()
```

```python
df.query('Region == "SEA"').groupby('Country or territory name')\
         ['Estimated prevalence of TB (all forms) per 100 000 population'].agg(['mean']) \
     .sort_values('mean')['mean'] \
    .iloc[0:10] \
    .plot(kind='barh', figsize=(12, 5), title='Top 10 Southeast Asia Countries for Estimated prevalence of TB per 100k')
ax.set_xlabel('#cases', 
               fontweight ='bold')
plt.show()
```
